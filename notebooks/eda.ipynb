```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../backend/creditcard.csv")
```

```python
data
```

```python
import seaborn as sns

# Class distribution
sns.countplot(x="Class", data=data)
plt.title("Class Distribution")
plt.show()
```

```python
# Distribution of transaction amounts
plt.figure(figsize=(10, 6))
sns.histplot(data[data["Class"] == 0]["Amount"], bins=50, color="blue", label="Non-Fraud", kde=True)
sns.histplot(data[data["Class"] == 1]["Amount"], bins=50, color="red", label="Fraud", kde=True)
plt.legend()
plt.title("Transaction Amount Distribution")
plt.show()
```

```python
# Convert 'Time' to hours for better interpretation
data["Hour"] = data["Time"] / 3600

# Plot fraud vs non-fraud over time
plt.figure(figsize=(10, 6))
sns.histplot(data[data["Class"] == 0]["Hour"], bins=48, color="blue", label="Non-Fraud", kde=True)
sns.histplot(data[data["Class"] == 1]["Hour"], bins=48, color="red", label="Fraud", kde=True)
plt.legend()
plt.title("Transaction Time Distribution")
plt.show()
```
